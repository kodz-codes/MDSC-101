```python
def matrix_input(mat_a, mat_b):
    #Taking the input values of matrix A
    for i in range(0,rows_of_mat_a):
        temp = []
        for j in range(0,cols_of_mat_a):
            print("please enter a number into the matrix positioned "+ str(i) +"th row and "+ str(j) +"th column")
            val = int(input())
            temp.append(val)
        mat_a.append(temp)
        
    print(mat_a)

    #Taking the input values of matrix B
    for z in range(0,rows_of_mat_b):
        temp = []
        for w in range(0,cols_of_mat_b):
            print("please enter a number into the matrix positioned "+ str(z) +"th row and "+ str(w) +"th column")
            val = int(input())
            temp.append(val)
        mat_b.append(temp)
        
    print(mat_b)
    return

def matrix_sub(mat_a,mat_b,result):
    for i in range(0,len(mat_a)):
        temp = []
        if (len(mat_a) == len(mat_b)) and (len(mat_a[i]) == len(mat_b[i])):
            #print("starting the matrix addition : ")
            for j in range(0,len(mat_a[i])):
                #print(i,j)
                val = mat_a[i][j] - mat_b[i][j]
                temp.append(val)
            result.append(temp)
            #print(mat_a[i][j])
        else: 
            print(" The dimensions entered for the matrix are not same to perform the matrix subtraction operation \n")
            return False
    return(result)

def matrix_add(mat_a,mat_b,result):
    for i in range(0,len(mat_a)):
        temp = []
        if (len(mat_a) == len(mat_b)) and (len(mat_a[i]) == len(mat_b[i])):
            #print("starting the matrix addition : ")
            for j in range(0,len(mat_a[i])):
                #print(i,j)
                val = mat_a[i][j] + mat_b[i][j]
                temp.append(val)
            result.append(temp)
            #print(mat_a[i][j])
        else: 
            print(" The dimensions entered for the matrix are not same to perform the matrix addition operation \n")
            return False
    return(result)

def matrix_mul(mat_a,mat_b,result):
    if (len(mat_a[0]) == len(mat_b)):
        for i in range(0,len(mat_a)):   # iterating through the rows of A
            temp = [] 
            for j in range(0,len(mat_b)):   # iterating through the columns of B
                #print(i,j)
                temp_val = 0
                for k in range(0,len(mat_a[i])):
                    val = mat_a[i][k] * mat_b[k][j]
                    print(mat_a[i][k],mat_b[k][j],temp_val)
                    temp_val = temp_val + val
                    print(temp_val)
                temp.append(temp_val)
            result.append(temp)
            #print(mat_a[i][j])
    else: 
        print(" The dimensions of the matrix are not appropriate to perform the matrix multiplication operation \n")
        return False
    return(result)




#Taking the dimensions of the matrices
rows_of_mat_a,cols_of_mat_a  = int(input("Please enter the no. of rows of matrix A : ")),int(input("Please enter the no. of columns of matrix  A : "))

rows_of_mat_b, cols_of_mat_b  = int(input("Please enter the no. of rows of matrix B : ")),int(input("Please enter the no. of columns of matrix B : "))

mat_a,mat_b,result = [],[],[] # intializing the mat_a, mat_b to null.

matrix_input(mat_a,mat_b) # taking the matrix input through a function.






#Getting the user input for a specific operation on matrices
user_input = int(input("PLease enter the number of the operation you want to perform : (e.g: 1 for Matrix addition and so on...) \n1.Matrix Addition \n2.Matrix Subtraction \n3.Matrix Multiplication \n"))

if(user_input == 1):
    output = matrix_add(mat_a,mat_b,result) # receiving the output from the matrix_add() function
    if type(output) == bool:
        print("Could not perform matrix operation due to the above reason")
    else: 
        print("After performing the matrix operation: ")
    print(output)
elif(user_input == 2):
    output = matrix_sub(mat_a,mat_b,result) # receiving the output from the matrix_sub() function
    if type(output) == bool:
        print("Could not perform matrix operation due to the above reason")
    else: 
        print("After performing the matrix operation: ")
    print(output)
elif(user_input == 3):
    output = matrix_mul(mat_a,mat_b,result) # receiving the output from the matrix_mul() function
    if type(output) == bool:
        print("Could not perform matrix operation due to the above reason")
    else: 
        print("After performing the matrix operation: ")
    print(output)
else:
    print("The input entered by the user is invalid...")


```
